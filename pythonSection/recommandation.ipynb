```python
import pandas as pd
from mlxtend.frequent_patterns import association_rules, apriori
```

```python
sales_receipts = pd.read_csv("../dataset/201904 sales reciepts.csv")
sales_receipts
```

```python
product = pd.read_csv("../dataset/product.csv")
product
```

```python
data = pd.merge(sales_receipts[['transaction_id','transaction_date', 'sales_outlet_id', 'customer_id', 'product_id', 'quantity']], product[['product_id', 'product_category', 'product']], on="product_id", how="left")
data 
```

```python
data['product'] = data['product'].str.replace(' Rg', '')
data['product'] = data['product'].str.replace(' Lg', '')
data['product'] = data['product'].str.replace(' Sm', '')
data
```

```python
data['product'].nunique()
```

```python
products_to_take = ['Cappuccino', 'Latte', 'Espresso shot',  \
                     'Dark chocolate','Sugar Free Vanilla syrup', 'Chocolate syrup',\
                    'Carmel syrup', 'Hazelnut syrup', 'Ginger Scone',  \
                    'Chocolate Croissant', 'Jumbo Savory Scone', 'Cranberry Scone', 'Hazelnut Biscotti',\
                    'Croissant', 'Almond Croissant', 'Oatmeal Scone', 'Chocolate Chip Biscotti',\
                    'Ginger Biscotti',\
                   ]
```

```python
data = data[data["product"].isin(products_to_take)] 
data
```

```python
data['product'].nunique()
```

```python
data[['product','product_category']].drop_duplicates().reset_index(drop=True)
```

```python
data['transaction'] = data['transaction_id'].astype(str) + "_" + data['customer_id'].astype(str)
data
```

```python
number_of_items_in_each_transaction = data['transaction'].value_counts().reset_index()
number_of_items_in_each_transaction
```

```python
valid_transactions = number_of_items_in_each_transaction[(number_of_items_in_each_transaction['count']>1)]['transaction'].tolist()
data = data[data['transaction'].isin(valid_transactions)]
data
```
